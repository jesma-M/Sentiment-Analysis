```python
import re
```

```python
def clean_html(text):
    clear=re.compile('<.*?>')
    return re.sub(clear,'',text)
```

```python
data['review']=data['review'].apply(clean_html)
```

```python
data
```

```python
data['review'].iloc[6]
```

```python
def remove_special(text):
    x=' '
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x        
```

```python
data['review']=data['review'].apply(remove_special)
```

```python
data['review']
```

```python
def convert_lower(text):
    return text.lower()
```

```python
data['review']=data['review'].apply(convert_lower)
```

```python
data['review']
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
a=stopwords.words('english')
```

```python
def remove_words(text):
    x=[]
    for i in text.split():
        if i not in a:
            x.append(i)
        else:
            pass
    return x        
```

```python
data['review']=data['review'].apply(remove_words)
```

```python
data
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem_words(text):
    x=[]
    for i in text:
        x.append(ps.stem(i))
    return x    
```

```python
data['review']=data['review'].apply(stem_words)
```

```python
data['review'].iloc[5]
```

```python
def join(text):
    return ' '.join(text)
```

```python
data['review']=data['review'].apply(join)
```

```python
data.iloc[5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
X=data['review']
y=data['sentiment']
```

```python
cv=CountVectorizer()
```

```python
cv.fit_transform(data['review'])
```

```python
X=cv.fit_transform(data['review']).toarray()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred=gnb.predict(X_test)
```

```python
gnb.score(X_test,y_test)
```

```python
ac=accuracy_score(y_test,y_pred)
```

```python
ac
```

```python
cm=confusion_matrix(y_test,y_pred)
```
